The dataset provided focuses on used car sales, capturing various attributes to analyze the second-hand car
market. This dataset provides insights into factors affecting resale value, trends in the used car industry, and
consumer behavior.

Description:

The dataset contains details about used cars listed for sale, such as brand, model, selling price, kilometers
driven, fuel type, and transmission type. This information is valuable for predictive modeling, market analysis,
and understanding customer preferences.

Features
 Car_id: A unique identifier for each car in the dataset, helping to track individual car entries
 Date: The date when the car sale transaction took place, formatted as YYYY-MM-DD
 Customer Name: The name of the customer who purchased the car, represented as a string
 Gender: The gender of the customer, categorized as "Male" or "Female.
 Annual Income: The customer's annual income in US dollars, represented as a numeric value
Dealer_Name: The name of the dealership selling the car, represented as a string
Company: The manufacturer or brand name of the car, such as "Toyota," "Ford," etc
 Model: The specific model name of the car, such as "Corolla," "Civic," etc
 Engine: The engine type of the car, such as "V6," "I4," etc
 Transmission: The type of transmission in the car, either "Manual" or "Automatic.
 Color: The color of the car, represented as a string (e.g., "Red," "Blue")
 Price ($): The selling price of the car in US dollars
 Dealer_No: A unique identifier for each car dealer in the dataset
 Body Style: The body style of the car, such as "Sedan," "SUV," etc
 Phone: The phone number of the customer who purchased the car
 Dealer_Region: The geographical region of the car dealer, such as "North," "South," etc.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv("Car Sale.csv")
df
```

```python
df.info()
```

```python
# What is the average selling price of cars for each dealer, and how does it compare across different dealers
avg_selling_price_dealer = df.groupby(["Dealer_Name"])["Price ($)"].mean().round(2)
print(f"Average selling price is {avg_selling_price_dealer}")
sns.histplot(avg_selling_price_dealer,bins=10)
```

```python
# Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing trends
df.columns
min_var = df.groupby("Company")["Price ($)"].min()
max_var = df.groupby("Company")["Price ($)"].max()

# find the min and max for each car brand and whichever got highest variation it is the answer
highest_var_prices = max_var - min_var
print(f"{highest_var_prices.idxmax()} has the highest variation i.e., {highest_var_prices.max()}")
```

```python
#What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare
df.columns
df["Transmission"].unique()
# to calculate the IQR for each transmission
iqr_data = df.groupby("Transmission")["Price ($)"].describe()[["25%","75%"]]
iqr_data["IQR"] = iqr_data["75%"] - iqr_data["25%"]
print("\nInterquartile Ranges (IQR) for each Transmission Type:\n", iqr_data)

ax=iqr_data["IQR"].plot(kind="bar",color=["skyblue","orange"])
plt.title("IQR of Auto and Manual")

for v in ax.patches:  
    plt.text(v.get_x() + v.get_width() / 2,  # X position (center of the bar)
             v.get_height() + 100,           # Y position (above the bar)
             f"{v.get_height():,.0f}",       # Text (formatted IQR value)
             ha="center", fontsize=12, color="black")
plt.show()

sns.boxplot(x=df["Transmission"],y=df["Price ($)"])
plt.title("Car proce distribution by transmission type")
plt.grid(True)
plt.show()
```

```python
# What is the distribution of car prices across different regions
df.columns
sns.boxplot(x=df["Dealer_Region"],y=df["Price ($)"],palette="pastel") # it is uniform
```

```python
# What is the distribution of cars based on body styles
body_style = df["Body Style"].value_counts()
sns.barplot(x=body_style.index, y=body_style.values,palette="pastel")
plt.xlabel("Body Style")
plt.ylabel("Number of Cars")
plt.title("Distribution of Cars by Body Style")

# to show values at top
for i,v in enumerate(body_style.values):
    plt.text(i,v+10,str(v),ha="center",fontsize=12,color="black")
plt.xticks(rotation=10)
plt.show()
```

```python
# How does the average selling price of cars vary by customer gender and annual income
df.columns
df["Annual Income"]
sns.scatterplot(x=df["Annual Income"],y=df["Price ($)"],hue=df["Gender"],alpha=0.7)

plt.xlabel("Annual Income")
plt.ylabel("Car Price ($)")
plt.title("Car Price Variation by Gender & Annual Income")

plt.xticks(rotation=30)  # Rotate if needed
plt.show()
```

```python
# What is the distribution of car prices by region, and how does the number of cars sold vary by region
plt.figure(figsize=(12, 6))
sns.boxplot(x=df["Dealer_Region"], y=df["Price ($)"], palette="Set2")

plt.xlabel("Region")
plt.ylabel("Car Price ($)")
plt.title("Distribution of Car Prices by Region")
plt.xticks(rotation=30)  # Rotate for better readability
plt.show()

# cars sold by region
plt.figure(figsize=(10,8))
sns.barplot(df["Dealer_Region"],palette="coolwarm")
plt.xlabel("Region")
plt.ylabel("Number of cars sold")
plt.title("No. of cars sold by region")
plt.xticks(rotation=10)
plt.show()
```

```python
# How does the average car price differ between cars with different engine sizesE
df.columns
avg_price_cars_diff_engine = df.groupby("Engine")["Price ($)"].mean().round(2)
avg_price_cars_diff_engine
sns.barplot(x=avg_price_cars_diff_engine.index,y=avg_price_cars_diff_engine.values,palette="coolwarm")
for i,v in enumerate(avg_price_cars_diff_engine.values):
    plt.text(i,v+10,str(v),ha="center",fontsize=12,color="black")
plt.xticks(rotation=10)
plt.show()
```

```python
# How do car prices vary based on the customer’s annual income bracketE
# df["Annual Income"] = df["Annual Income"] * 100000
sns.scatterplot(x=df["Annual Income"],y=df["Price ($)"],alpha=0.5)
plt.xscale("log")
plt.xlabel("Annual income")
plt.ylabel("prices")
plt.title("Annual Income vs Car Price")
plt.show()
```

```python
#  What are the top 5 car models with the highest number of sales, and how does their price distribution lookE
df.columns
top_5 = df["Model"].value_counts().nlargest(5).index
print(f"Top 5 Model {top_5}")

# in order to find price distribution lets create dataframe for this only
top_5_prices = df[df["Model"].isin(top_5)]
sns.boxplot(x="Model",y="Price ($)",data=top_5_prices,palette="coolwarm")
plt.xlabel("Car Model")
plt.ylabel("Price ($)")
plt.title("Price Distribution of Top 5 Best-Selling Car Models")
plt.xticks(rotation=45)  # Rotate labels if needed
plt.show()
```

```python
# How does car price vary with engine size across different car colors, and which colors have the highest price
# variationE
df.columns
df.Engine.unique()
sns.barplot(x=df["Engine"],y=df["Price ($)"],hue=df["Color"],palette="Set2")
plt.xlabel("Engine Type")
plt.ylabel("Average Car Price ($)")
plt.title("Car Price Variation by Engine Type and Color")
plt.show()

#  Color with the Highest Price Variation
price_var = df.groupby("Color")["Price ($)"].agg(["std",lambda x:x.quantile(0.75)-x.quantile(0.25)])
price_var.columns = ["Std","IQR"]
print(price_var.sort_values("IQR",ascending=False))
```

```python
#  Is there any seasonal trend in car sales based on the date of sale
df.columns
df["Date"] = pd.to_datetime(df["Date"])
df["month"] = pd.to_datetime(df["Date"]).dt.month
df["year"] = pd.to_datetime(df["Date"]).dt.year
year = df["year"].value_counts()
month=df["month"].value_counts()
sns.barplot(x=month.index,y=month.values)
plt.show()
# november and december month has the highest sales
sns.barplot(x=year.index,y=year.values)
plt.show()
```
