Description:

This dataset contains information on 1K+ Amazon products, including their ratings, reviews, and other details
Features:

product_id: Unique identifier for each product

product_name: Name of the product

category: Category of the product

discounted_price: Discounted price of the product

actual_price: Actual price of the product

discount_percentage: Percentage of discount for the product

rating: Rating of the product (1-5)

rating_count: Number of people who voted for the Amazon rating

about_product: Description about the product

user_id: ID of the user who wrote the review

user_name: Name of the user who wrote the review

review_id: ID of the user review

review_title: Short review

review_content: Long review

img_link: Image link of the product

product_link: Official website link of the product

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv("amazon.csv")
df
```

```python
df.columns
```

```python
len(df["category"].unique())
```

```python
df["rating"] = df["rating"].replace("|",1)
df["rating"] = df["rating"].astype(float)
```

```python
df["rating"].unique()
```

```python
# What is the average rating for each product category
# Calculate average rating for each product category
category_avg_rating = df.groupby("category")["rating"].mean().round(2)

# Display the result
print(category_avg_rating)

```

```python
# What are the top rating_count products by category
top_20_products=category_avg_rating.sort_values(ascending=False).head(20)
top_20_products
```

```python
df.columns
```

```python
df.dtypes
```

```python
df["actual_price"] = df["actual_price"].astype(str).str.extract(r"([\d,.]+)")[0].str.replace(",", "", regex=True).astype(float)
df["discounted_price"] = df["discounted_price"].astype(str).str.extract(r"([\d,.]+)")[0].str.replace(",", "", regex=True).astype(float)
df["discount_percentage"] = df["discount_percentage"].astype(str).str.extract(r"([\d,.]+)")[0].str.replace(",", "", regex=True).astype(float)
```

```python
# What is the distribution of discounted prices vs. actual pricesk
# Scatter plot
plt.figure(figsize=(10, 6))
plt.hexbin(df["actual_price"], df["discounted_price"], gridsize=50, cmap="Blues", alpha=0.8)

plt.colorbar(label="Count of Data Points")
plt.xlabel("Actual Price")
plt.ylabel("Discounted Price")
plt.title("Hexbin Plot of Discounted vs. Actual Prices")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=df["actual_price"], y=df["discounted_price"])
plt.xlabel("Actual Price")
plt.ylabel("Discounted Price")
plt.title("Box Plot of Discounted Prices vs. Actual Prices")
plt.show()
```

# as the actual prize increases discount also increases

```python
df["discount_percentage"]
```

```python
# How does the average discount percentage vary across categories
# Group by category and calculate the average discount percentage
# df["discount_percentage"] = pd.to_numeric(df["discount_percentage"], errors="coerce")

category_discount = df.groupby("category")["discount_percentage"].mean().reset_index()

# Sort categories by discount percentage
category_discount = category_discount.sort_values(by="discount_percentage", ascending=False)

# Plot the results
plt.figure(figsize=(12, 6))
sns.barplot(data=category_discount, x="category", y="discount_percentage", palette="coolwarm")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Category")
plt.ylabel("Average Discount (%)")
plt.title("Average Discount Percentage Across Categories")
plt.show()
```
