Description of the Dataset:

The dataset titled "Spotify Data: Popular Hip-hop Artists and Tracks" provides a curated collection of
approximately 500 entries showcasing the vibrant realm of hip-hop music. These entries meticulously compile
the most celebrated hip-hop tracks and artists, reflecting their significant influence on the genre's landscape.
Each entry not only highlights the popularity and musical composition of the tracks but also underscores the
creative prowess of the artists and their profound impact on global listeners.


Application in Data Science:

This dataset serves as a valuable resource for various data science explorations. Analysts can delve into trend
analysis to discern the popularity dynamics of hit hip-hop tracks over recent years. Additionally, the dataset
enables network analysis to uncover collaborative patterns among top artists, shedding light on the genre's
evolving collaborative landscape. Furthermore, it facilitates the development of predictive models aimed at
forecasting track popularity based on diverse features, offering insights for artists, producers, and marketers.
Column Descriptors:


Artist: The name of the artist, providing direct attribution to the creative mind behind the track.

Track Name: The title of the track, encapsulating its identity and essence.

Popularity: A numeric score reflecting the track's reception and appeal among Spotify listeners.

Duration (ms): The track's length in milliseconds, detailing the temporal extent of the musical experience.

Track ID: A unique identifier within Spotify's ecosystem, enabling direct access to the track for further
exploration.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv("spotify.csv")
df
```

```python
df.drop_duplicates()
```

```python
#What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram

sns.histplot(df["Popularity"],bins=20,color="green")
plt.xlabel('Popularity')
plt.ylabel('Number of Tracks')
plt.title('Distribution of Track Popularity')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

```python
# Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot
sns.scatterplot(x="Popularity",y="Duration (ms)",data=df,alpha=0.5)
plt.title("R/S between popularity and Duration")
plt.show()
# Hence there is r/s between popularity and duration as one increases other increases
```

```python
# Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using
#a countplot
artist_count = df["Artist"].value_counts()
top_artist = artist_count.head(10)
top_artist
sns.barplot(x=top_artist.index,y=top_artist.values,palette="coolwarm")
plt.xlabel("Artist Name")
plt.ylabel("Count")
plt.title("Top 10 Artist with highest no. of tracks")
plt.grid(axis="y",linestyle="--")
plt.xticks(rotation=45)
plt.show()
```

```python
# What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each
artist_least_count = artist_count.tail(5)
artist_least_count
```

```python
# Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and
#display the average popularity for each artist
top_five = artist_count.head(5)
top_five
for index, value in top_five.items():
    avg_pop = df[df["Artist"]==index]["Popularity"].mean().round(2)
    print(f"The Average Popularity of {index} is {avg_pop}")
    
```

```python
# For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist
for index, value in top_five.items():
    most_pop_track = df[df["Artist"]==index].sort_values(by="Popularity",ascending=False).iloc[0]["Track Name"]
    print(f"The most popular Track of {index} is {most_pop_track}")
```

```python
#  Visualize relationships between multiple numerical variables simultaneously using a pair plot
numerical_cols = ["Popularity", "Duration (ms)"]  # Add more if available

# Creating the pair plot
sns.pairplot(df[numerical_cols], diag_kind='kde', plot_kws={'alpha': 0.6})

# Show the plot
plt.show()
```

```python
# Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or
# violin plot
plt.figure(figsize=(15, 6))
sns.boxplot(x="Artist", y="Duration (ms)", data=df, palette="coolwarm")

# Labels and title
plt.xlabel("Artist")
plt.ylabel("Duration (ms)")
plt.title("Variation of Track Duration Across Artists")
plt.xticks(rotation=90)  # Rotate artist names for readability

# Show plot
plt.show()
```

```python
# How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a
#violin plot.
# Filter artists with at least 5 tracks to avoid clutter
artist_counts = df['Artist'].value_counts()
selected_artists = artist_counts[artist_counts >= 5].index
filtered_df = df[df['Artist'].isin(selected_artists)]

# Create a swarm plot
plt.figure(figsize=(15, 6))
sns.swarmplot(x="Artist", y="Popularity", data=filtered_df, palette="dark")

# Labels and title
plt.xlabel("Artist")
plt.ylabel("Popularity")
plt.title("Distribution of Track Popularity Across Artists")
plt.xticks(rotation=90)  # Rotate artist names for readability

# Show plot
plt.show()
```
